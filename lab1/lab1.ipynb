# PIL + matplotlib

```python
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
```

```python
# 1. đọc ảnh
img = Image.open(r"D:\xu_li_anh\lab2\Voi.jpg")
```

```python
w,h =img.size
print(w)
print(h)
```

```python
img.mode
```

```python

plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()

# 2. lưu ảnh định dạng khác
img.save("output.png")
```

```python
# 3. chuyển sang grayscale
gray = img.convert("L")

plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# 4. crop ảnh
crop = img.crop((50, 50, 2000, 2000))

plt.imshow(crop)
plt.title("Cropped Image")
plt.axis("off")
plt.show()
```

```python
# 5. resize ảnh
resize = img.resize((500,500))

plt.imshow(resize)
plt.title("Resized Image")
plt.axis("off")
plt.show()
```

```python
# 6. vẽ hình
img_draw = img.copy()
draw = ImageDraw.Draw(img_draw)

draw.rectangle((100,100,1100,1100), outline="red", width=3)
draw.line((0,0,1300,1900), fill="blue", width=3)

plt.imshow(img_draw)
plt.title("Draw Shapes")
plt.axis("off")
plt.show()
```

```python
# 7. thêm text
font = ImageFont.truetype("arial.ttf", 150)
draw.text((780,180), "NHóm 4", fill="red", font=font)
plt.imshow(img_draw) 
plt.title("Image with Text") 
plt.axis("off") 
plt.show()
```
