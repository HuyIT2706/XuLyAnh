```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt


# 1. đọc ảnh
img = Image.open(r"./assets/input.jpg")

plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()

# 2. lưu ảnh định dạng khác
img.save("./assets/output.png")

# 3. chuyển sang grayscale
gray = img.convert("L")

plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()

# 4. crop ảnh
crop = img.crop((50, 50, 200, 200))

plt.imshow(crop)
plt.title("Cropped Image")
plt.axis("off")
plt.show()

# 5. resize ảnh
resize = img.resize((200,200))

plt.imshow(resize)
plt.title("Resized Image")
plt.axis("off")
plt.show()

# 6. vẽ hình
img_draw = img.copy()
draw = ImageDraw.Draw(img_draw)

draw.rectangle((0,0,800,700), outline="red", width=50)
draw.line((1200,1210,1200,1200), fill="blue", width=50)

plt.imshow(img_draw)
plt.title("Draw Shapes")
plt.axis("off")
plt.show()

# 7. thêm text
img_draw = img.copy()
draw = ImageDraw.Draw(img_draw)
font = ImageFont.truetype("arial.ttf", 100)
draw.text((500,500),"Bui Van Huy", font=font, fill="red")

plt.imshow(img_draw)
plt.title("Image with Text")
plt.axis("off")
plt.show()
```
